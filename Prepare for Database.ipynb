```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import xml.etree.ElementTree as ET
import pprint
import re
import codecs
import json
```

```python
lower = re.compile(r'^([a-z]|_)*$')
lower_colon = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')

CREATED = [ "version", "changeset", "timestamp", "user", "uid"]
valid_cities = ['Apex','Morrisville','Raleigh','Cary']
```

```python
street_mapping = { "St": "Street",
            "St.": "Street",
            "Ave" : "Avenue",
            "Rd." : "Road",
            "Rd" : "Road",
            "Cir" : "Circle",
            "Cir." : "Circle",
            "Pl" : "Place",
            "Pl." : "Place",
            "Dr" : "Drive",
            "Dr." : "Drive"
            
            }
def update_street_name(name, mapping):

    for k, v in mapping.items():        
        match = re.search(k+"$",name)
        if match:
            name = re.sub(k,v,name)
            #print "Pattern:|%s| Replacement:|%s| New String:|%s|" % (k,v,name)
            break

    return name
```

```python
def scrub_city(i):
    
    if i not in valid_cities:
        for valid in valid_cities: 
            if re.search(valid,i,re.IGNORECASE): i = valid

    return (i)
```

```python
#
# Phone number regex 
#
#Gold Standard: +1 919 555 1212
phoneformat = re.compile(r'\+1 \d\d\d \d\d\d \d\d\d\d')
#Ex: 919 5551212 or 919-5551212
badphoneformat = re.compile(r'\d\d\d\W\d\d\d\d\d\d\d')
#Ex: (919) 555-1212 or (919) 555 1212
badphoneformat2 = re.compile(r'\(\d\d\d\)\W\d\d\d\W\d\d\d\d')
#Ex: 919-555-1212
badphoneformat3 = re.compile(r'\d\d\d\W\d\d\d\W\d\d\d\d')
#Ex: 9195551212
badphoneformat4 = re.compile(r'\d\d\d\d\d\d\d\d\d\d')
```

```python
def Phone_Processing(node,key,value):
#
# Any tags that have the phrase 'phone' or 'fax' in the key are scrubbed to match a gold standard
# of a phone number format for US.  
#
    if re.search(badphoneformat,value):
        node[key] = "+1 "+value[-11:-8]+" "+value[-7:-4]+" "+value[-4:]
    elif re.search(badphoneformat2,value):
        node[key] = "+1 "+value[-13:-10]+" "+value[-8:-5]+" "+value[-4:]
    elif re.search(badphoneformat3,value):
        node[key] = "+1 "+value[-12:-9]+" "+value[-8:-5]+" "+value[-4:]
    elif re.search(badphoneformat4,value):
        node[key] = "+1 "+value[-10:-7]+" "+value[-7:-4]+" "+value[-4:]
    else: node[key] = value    

    return(node)
```

```python
def Address_Processing(node,key,value):
#
# Any tags that have k values beginning with 'addr:' are processed and placed in a dictionary within node
# with a key value of "address"  Types of processing:
#  -Street Types are harmonized.  
#
 #Initialize dictionary with key labeled "address" in node dictionary if not yet created
    if "address" not in node:
        node["address"] ={}
    #
    # Street Processing
    #
    if key == "addr:street":
        better_name = update_street_name(value, street_mapping)
        node["address"][key[5:]] = better_name
        #print "Street: ", node["address"][key[5:]]
    #
    # City Name Processing
    #
    elif key == "addr:city":
        better_name = scrub_city(value)
        node["address"][key[5:]] = better_name

    elif key == "addr:state": node["address"]["state"] = "NC"
    elif key == "addr:country": node["address"]["country"] = "US"   
    #
    # Fix some known bad records found in audit
    #
    elif key == "addr:housename" and value == "Suite 160":
        node["address"]["suite"] = "160"
    elif key == "addr:full" and re.search(r"Avenue Suite",value):
        node["address"]["full"] = re.sub(r"Avenue Suite","Avenue, Suite",value)
        
    #    
    # Catch all for other types of addr: types
    #
    else:
        
        node["address"][key[5:]] = value
        #print "Catch All:", node["address"][key[5:]]
            
    return(node)
```

```python
def shape_element(element):

    #All aspects of valid OSM elements will be put into the node dictionary.  Init node dict here
    node = {}
    
    #Only process node or way OSM elements
    if element.tag == "node" or element.tag == "way" :  
        waynodes = []
        if element.tag == "node": node["type"] = "node"
        else: node["type"] = "way"
        for elem in element.iter():

            #
            # ID Line Processing
            #
            if ("id" in elem.attrib):   #We have the first line of the element
                #First Line Processing
                pos = [None,None] #initialize position array for lat & long
                
                
                #Certain attributes of the first line in the OSM element will be placed in embedded dictionary of node
                #with a key of "created".  The attributes to be placed are defined in list CREATED
                for topkey in elem.attrib:
                    if topkey in CREATED: 
                        if "created" not in node:
                            node["created"] = {}
                        node["created"][topkey] = elem.attrib[topkey]
                    
                    # Latitutde and Longitude are placed in a placeholder array
                    elif topkey == "lat": pos[0] = float(elem.attrib[topkey])
                    elif topkey == "lon" : pos[1] = float(elem.attrib[topkey])
                    else: node[topkey] = elem.attrib[topkey]
                        
                # If placeholder array is fully populated, set latitude & longitude array  to key "pos"
                if pos[0] is not None and pos[1] is not None:  node["pos"] = pos  
                    
            #
            # TAG processing for both nodes & ways - get attributes of k & v
            #
            elif ("k" in elem.attrib) and ("v" in elem.attrib) and (elem.tag == "tag"):  #We have a key value tag
                k,v = elem.attrib["k"],elem.attrib["v"]
                #print "Analyzing key: |%s| value: |%s|" %(k,v)
                
                #
                # Don't let bad python keys cause problems
                #
                if problemchars.search(k): 
                    print "PROBLEM Characters!!!!!!",k,v
                #
                # Special processing for any keys starting with "addr:"
                #    
                elif re.search(r'^addr:',k):
                    if not re.search(r'addr:street:',k):  #Strip out any addr:street entries that have a second ':'
                        
                        node = Address_Processing(node,k,v)
                #
                # Special processing for phone numbers
                #
                elif re.search(r'phone',k) or re.search(r'fax',k):
                    node = Phone_Processing(node,k,v)
                else: node[k] = v
            
            
            #
            # Way Processing - Build array of node referrences
            #
            elif (element.tag == "way") and (elem.tag == "nd"): 
                waynodes.append(elem.attrib["ref"])
            
            #
            # Catch the unexpected for Nodes or Ways
            #
            else: 
                print "%s Did Not Match Analysis" % elem.tag    
                print "Attributes are:"
                pprint.pprint(elem.attrib)
        
        # For ways, create a key labeled "node_refs" and provide it the array of the referenced nodes
        if (len(waynodes) <> 0) and (element.tag == "way"): node["node_refs"] = waynodes
        
        
        ### Debug Help####
        #print "##### JSON for %s id: %s ######" % (element.tag, element.attrib["id"])
        #pprint.pprint(node)
        #print "###### Node Processing Complete ###"
        return node
    else:
        return None
```

```python
def process_map(file_in, pretty = False):
    # You do not need to change this file
    eventnum = 2
    file_out = "{0}.json".format(file_in)

    data = []
    with codecs.open(file_out, "w") as fo:
        context = iter(ET.iterparse(file_in,events=('start','end')))
        _,root = next(context)
      
        for event, element in context:


            if event == 'start':
                eventnum += 1
                #print  "EvenNum: %i  Tag: %s %s" %( eventnum,element.tag,element.get("v"))
                el = shape_element(element)
                if el:
                    data.append(el)
                    if pretty: 
                        fo.write(json.dumps(el, indent=2)+"\n")
                    else:
                        fo.write(json.dumps(el) + "\n")
            elif event == 'end':
                root.clear()

    return data
```
